**01. CSV 파일 불러오기 및 Data 확인** 
- 연희동 인구 데이터

```python
#1) 필요한 라이브러리 불러오기
import matplotlib.pyplot as plt
import csv
#2) 인구 데이터 불러오기
f = open('population.csv', encoding='euc-kr')
data = csv.reader(f)
#3) 우리 동네 인구 데이터를 연령 순서대로 저장하기
mytown=[]
for row in data :
  if '연희동' in row[0] :
    for i in range(3, 104) :
      mytown.append(int(row[i]))
print(mytown)
```

**02. 데이터 시각화** - plot() 함수 
- 연희동의 인구 구조 데이터 시각화

```python
#1) 필요한 라이브러리 불러오기
import matplotlib.pyplot as plt
import csv
#2) 인구 데이터 불러오기
f = open('population.csv', encoding='euc-kr')
data = csv.reader(f)
#3) 우리 동네 인구 데이터를 연령 순서대로 저장하기
mytown = []
for row in data:
  if '연희동' in row[0]:
    for i in range(3, 104):
      mytown.append(int(row[i]))
print(mytown)
#4) 저장된 데이터를 꺾은선 그래프로 그리기
plt.figure(dpi = 150) #그래프 퀄리티
plt.style.use('ggplot') #그래프 스타잍 변경
plt.rc('font', family='NanumBarunGothic') #matplotlib의 폰트를 나눔 고딕 폰트로 지정
plt.title('연희동의 인구 구조')
plt.plot(mytown)
plt.show()
```
