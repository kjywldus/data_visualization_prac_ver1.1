**02. Data 시각화 응용** 
- 제주도에는 정말 여성 인구가 더 많을까?

```python
#1) 필요한 라이브러리 불러오기
import matplotlib.pyplot as plt
import csv
#2) 성별 인구 데이터 불러오기
f = open('gender.csv', encoding='euc-kr')
data = csv.reader(f)
m = []
f = []
#3-1) 남성, 여성의 데이터를 연령 순으로 저장하기
for row in data :
  if '제주' in row[0]:
    for i in range(3,104) :
      m.append(-int(row[i]))
      f.append(int(row[i+103]))
    break
#4) 저장된 데이터를 막대 그래프로 그리기
plt.figure(dpi = 100)
plt.style.use('ggplot')
plt.rc('font', family='NanumBarunGothic')
plt.title('제주도의 성별 인구 구조')
plt.barh(range(101), m, label = '남')
plt.barh(range(101), f, label = '여')
plt.rcParams['axes.unicode_minus'] = False
plt.legend()
#5) 결과 그래프 보여주기
plt.show()
```
