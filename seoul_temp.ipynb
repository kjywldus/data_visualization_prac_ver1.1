```python
import matplotlib.pyplot as plt
import csv
f = open('seoul_temp.csv',encoding='euc-kr')
data = csv.reader(f)
next(data)
mx_temp = []
for row in data:
  if row[-1] != '' and row[0].split('-')[1] == '11'and row[0].split('-')[2] == '16':
    mx_temp.append(float(row[-1]))
plt.title('11-16 year temperature')
plt.ylabel('Temperature(C)')
plt.plot(mx_temp, color = 'r')
plt.show()
```
