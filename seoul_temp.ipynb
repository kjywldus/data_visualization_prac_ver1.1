**03. 서울 기온 데이터 시각화 적용 - plot() 함수**
- 1907년부터 2022년 21일까지의 서울 기온 데이터 시각화

```python
import matplotlib.pyplot as plt
import csv
f= open('seoul_temp.csv', encoding='euc-kr')
data = csv.reader(f)
next(data)
mx_temp = []
for row in data:
  if row[-1] != '' :
    mx_temp.append(float(row[-1]))
plt.plot(mx_temp)
plt.show()
```

**03. 서울 기온 데이터 시각화 적용 - split(), title(), ylable(), plot() 함수, 색상** 
- 1907년부터 2022년 12월 21일까지의 서울 기온 데이터에서 생일에 해당하는 월과 일의 기온 변화 시각화

```python
import matplotlib.pyplot as plt
import csv
f = open('seoul_temp.csv',encoding='euc-kr')
data = csv.reader(f)
next(data)
mx_temp = []
for row in data:
  if row[-1] != '' and row[0].split('-')[1] == '11'and row[0].split('-')[2] == '16':
    mx_temp.append(float(row[-1]))
plt.title('11-16 year temperature')
plt.ylabel('Temperature(C)')
plt.plot(mx_temp, color = 'r')
plt.show()
```
